```
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torch.optim as optim
import argparse
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
from tqdm import tqdm
from torchvision import datasets
from torch.utils.data import DataLoader
from pandas import DataFrame
import pandas as pd
from pandas import ExcelWriter
import time
from torchvision.utils import save_image
from matplotlib.backends.backend_pdf import PdfPages
from torchvision.utils import make_grid
from google.colab import drive
drive.mount('/content/drive/')
torch.manual_seed(0)

def show_tensor_images(image_tensor,title,num_images=25,size=(1,32,32)):
    image_tensor = (image_tensor+1)/2
    image_unflat = image_tensor.detach().cpu()
    image_grid = make_grid(image_unflat[:num_images],nrow=5)
    plt.figure()
    plt.title(title)
    plt.imshow(image_grid.permute(1,2,0).squeeze())

def multipage(filename,figs=None,dpi=200):
    pp = PdfPages(filename)
    if figs is None:
        figs = [plt.figure(n) for n in plt.get_fignums()]
    for fig in figs:
        fig.savefig(pp,format='pdf')
    pp.close()


class LinearVAE(nn.Module):
    def __init__(self):
        super(LinearVAE,self).__init__()
        self.enc1 = nn.Linear(in_features=3072, out_features=512)
        self.enc2 = nn.Linear(in_features=512, out_features=features*2)
        self.dec1 = nn.Linear(in_features=features, out_features=512)
        self.dec2 = nn.Linear(in_features=512, out_features=3072)

    def reparameterize(self,mu,log_var):
        std = torch.exp(0.5*log_var)
        eps = torch.randn_like(std)
        sample = mu + (eps*std)
        return sample

    def forward(self,x):
        x = F.relu(self.enc1(x))
        x = self.enc2(x).view(-1,2,features)
        mu = x[:,0,:]
        log_var = x[:,1,:]
        z = self.reparameterize(mu,log_var)
        x = F.relu(self.dec1(z))
        reconstruction = torch.sigmoid(self.dec2(x))
        return reconstruction, mu, log_var

class LinearVAE_2(nn.Module):
    def __init__(self):
        super(LinearVAE_2,self).__init__()
        self.enc1 = nn.Linear(in_features=3072, out_features=1024)
        self.enc2 = nn.Linear(in_features=1024, out_features=512)
        self.enc3 = nn.Linear(in_features=512, out_features=features*2)
        self.dec1 = nn.Linear(in_features=features, out_features=512)
        self.dec2 = nn.Linear(in_features=512, out_features=1024)
        self.dec3 = nn.Linear(in_features=1024, out_features=3072)

    def reparameterize(self,mu,log_var):
        std = torch.exp(0.5*log_var)
        eps = torch.randn_like(std)
        sample = mu + (eps*std)
        return sample

    def forward(self,x):
        x = F.relu(self.enc1(x))
        x = F.relu(self.enc2(x))
        x = self.enc3(x).view(-1,2,features)
        mu = x[:,0,:]
        log_var = x[:,1,:]
        z = self.reparameterize(mu,log_var)
        x = F.relu(self.dec1(z))
        x = F.relu(self.dec2(x))
        reconstruction = torch.sigmoid(self.dec3(x))
        return reconstruction, mu, log_var

def final_loss(bce_loss, mu, logvar):
    BCE = bce_loss
    KLD = -0.5*torch.sum(1+logvar-mu.pow(2)-logvar.exp())
    return BCE+KLD

def fit(model,dataloader):
    model.train()
    running_loss = 0
    load_num = 0
    for i, data in tqdm(enumerate(dataloader),total=int(len(train_data)/dataloader.batch_size)):
        data,_ = data
        data = data.to(device)
        data = data.view(data.size(0),-1)
        optimizer.zero_grad()
        reconstruction, mu, logvar = model(data)
        bce_loss = criterion(reconstruction, data)
        loss = final_loss(bce_loss, mu, logvar)
        running_loss += loss.item()
        loss.backward()
        optimizer.step()
        load_num += 1
    train_loss = running_loss/len(dataloader.dataset)
    print(f"dataloader dataset is: {len(dataloader.dataset)}")
    print(f"load_num is: {load_num}")
    return train_loss

def validate(model, dataloader,title):
    model.eval()
    running_loss = 0
    load_num = 0
    with torch.no_grad():
        for i, data in tqdm(enumerate(dataloader),total=int(len(val_data)/dataloader.batch_size)):
            data, _ = data
            data = data.to(device)
            data = data.view(data.size(0), -1)
            reconstruction, mu, logvar = model(data)
            bce_loss = criterion(reconstruction, data)
            loss = final_loss(bce_loss, mu, logvar)
            running_loss += loss.item()
            load_num += 1
            if (i == int(len(val_data)/dataloader.batch_size) - 1) and ((epoch+1)%2==0):
                num_rows = 8
                both = torch.cat((data.view(batch_size, 3, 32, 32)[:8],reconstruction.view(batch_size, 3, 32, 32)[:8]))
                save_image(both.cpu(),fp='/content/drive/My Drive/VAE_CIFAR10/(n_layers=3)real&fake MNIST image'+title+'.png',nrow=num_rows)
            '''
            if (i == int(len(val_data)/dataloader.batch_size) - 1) and (epoch%1==0):
                real = data.view(batch_size, 3, 32, 32)[:8]
                fake = reconstruction.view(batch_size, 3, 32, 32)[:8]
                show_tensor_images(real,str(epoch))
                show_tensor_images(fake,str(epoch))
            '''
    val_loss = running_loss/len(dataloader.dataset)
    return val_loss

#start to train:
start_time = time.time()
parser = argparse.ArgumentParser()
parser.add_argument('-e','--n_epochs',default=10,type=int,help='number of n_epochs to train the VAE for')
parser.add_argument("-f", "--fff", help="a dummy argument to fool ipython", default="1")
args = vars(parser.parse_args())
#n_epochs = args['n_epochs']
batch_size = 64
lr = 0.0001
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
transform = transforms.Compose([transforms.ToTensor(),])
train_data = datasets.CIFAR10(root='.',train=True,download=True,transform=transform)
val_data = datasets.CIFAR10(root='.',train=False,download=True,transform=transform)
train_loader = DataLoader(train_data,batch_size=batch_size,shuffle=True)
val_loader = DataLoader(val_data,batch_size=batch_size,shuffle=False)
train_loss = []
val_loss = []
data_list = []

n_epochs = 50
#features_list = [32]
features_list = [32,50,64,100]
n_layers=3

for features in features_list:
    model = LinearVAE_2().to(device)
    optimizer = optim.Adam(model.parameters(),lr=lr)
    criterion = nn.BCELoss(reduction='sum')
    #range(n_epochs)
    for epoch in range(n_epochs):
        print(f'Epoch {epoch+1} of {n_epochs}')
        train_epoch_loss = fit(model, train_loader)
        title='(n_layers='+str(n_layers)+',z_dim='+str(features)+',epoch'+str(epoch)+')'
        val_epoch_loss = validate(model, val_loader,title)
        data_list.append([epoch+1,features, train_epoch_loss, val_epoch_loss])
        train_loss.append(train_epoch_loss)
        val_loss.append(val_epoch_loss)
        print(f'epoch {epoch} of {n_epochs}, train loss: {train_epoch_loss:4f}, validation loss: {val_epoch_loss:.4f}')

    plt.figure()
    plt.title('VAE CIFAR10 loss (n_epochs='+str(n_epochs)+',z_dim='+str(features)+')')
    plt.plot(range(1,n_epochs+1),train_loss,color='blue',label='training_loss')
    plt.plot(range(1,n_epochs+1),val_loss,color='red',label='validation_loss')
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.legend()

    train_loss = []
    val_loss = []

multipage('/content/drive/My Drive/VAE_CIFAR10/VAE CIFAR10 record(n_epochs='+str(n_epochs)+',z_dim='+str(features)+')')

data_df = pd.DataFrame(data_list)
data_df.columns = ['epoch','z_dim','train_loss','validation_loss']

with ExcelWriter('/content/drive/My Drive/VAE_CIFAR10/VAE_CIFAR10_output_table.xlsx',engine='openpyxl',mode='a') as writer:
    data_df.to_excel(writer,index=False)
    writer.save()

total_time = time.time()-start_time
total_time = round(total_time/60,2)

with open('/content/drive/My Drive/VAE_CIFAR10/record for VAE CIFAR10.txt','a') as to_file:
    to_file.write('\n')
    to_file.write('n_layers=3')
    to_file.write('\n')
    to_file.write('VAE CIFAR10 record(n_epochs='+str(n_epochs)+',z_dim='+str(features)+')')
    to_file.write('\n')
    to_file.write('total runing time(minutes) is:'+str(total_time))
    to_file.close()


print(f'stop training, total time(minutes) is:{total_time}')
```
